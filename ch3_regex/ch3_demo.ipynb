```python
"問題20-28の流れを再現する"

# イギリスの記事を抽出
uk_article = extract_country(country_dfs, 'イギリス')

# イギリスの記事からカテゴリ名を抽出
uk_categories = extract_categories(uk_article)

# イギリスの記事からセクション構造を抽出
uk_sections = extract_sections(uk_article, view=False)

# イギリスの記事からファイル名を抽出
uk_files = extract_files(uk_article)

# イギリスの記事から基本情報を抽出
uk_main_info = extract_main_info(uk_article)

# イギリスの記事の基本情報から強調マークアップを除去する
uk_strong_removed = remove_strong(uk_main_info)

# さきほど強調マークアップを除去したものに対して，内部リンクを除去する
uk_inlink_removed = remove_inlink(uk_strong_removed)

uk_markups_removed = remove_markups(uk_inlink_removed)
display(dict2df(uk_markups_removed))
```
