# 19 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

```python
import pandas as pd
# データの読み込み
df = pd.read_table('popular_names.txt', header=None)
# 1列目だけ取り出す
names = df[0]
```

```python
# uniqueな要素をリストで取り出したいならこれでOK．ただし出現頻度は得られない
# print(names.unique())

# 出現頻度の降順で名前と出現回数のペアが得られる
rank = names.value_counts()
print(rank)
print(type(rank)) # => <class 'pandas.core.series.Series'>

# dict型にして出力
rank = dict(rank)
print(rank) # 名前がkeyで出現回数がvalのdictになっていて，valの降順に並んでいる
```

```python
# keyだけを順に出力していく
for key in rank.keys():
    print(key)
```

## コマンドで確認

`cut`，`uniq`，`sort`をうまく利用する．

まず `cut` コマンドで `popular_names.txt` のうち名前のフィールドである1列目を抜き出す.
次に，`sort` コマンドで名前を辞書順にソートした後に `uniq`コマンドで重複を取り除く．このとき，`-c` オプションをつけることで各名前の重複した個数も同時に出力してくれる．最後に，`sort` コマンドを用いてそれを出現回数の降順でソートする．数値の降順ソートにするために`-n`，`-r` オプションをつける．

以下のコマンドを実行すれば，`rank.txt` に出現回数の降順に出現回数と名前が出力される．

`cut -f 1 popular_names.txt | sort | uniq -c | sort -n -r > rank.txt`


- `cut -f 1` は1番目のフィールドを切り出すという意味だった．フィールド区切り文字は`-d`オプションで指定できるけど今回はしなくてもうまくいく（最初から空白文字で区切られているので）．
- 最後の `sort -n -r`　は `sort -nr` と略記可能．
- また，`-k`オプションで何列目でソートできるか指定できるけど多分デフォルト1になっているので今回なくてもうまくいく．
