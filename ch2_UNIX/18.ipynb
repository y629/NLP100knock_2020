# 18 各行を3コラム目の数値の降順にソート
各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．

```python
import pandas as pd
# データの読み込み
df = pd.read_table('popular_names.txt', header=None)

# 3列目(0-indexedなので2列目)の要素で降順ソート(昇順にしたかったらascending=Trueにせよ)
df = df.sort_values(2, ascending=False)
display(df)
```

## コマンドで確認

`sort`コマンドを用いれば特定のフィールドの値でソートできる．
ソートした後に最初の10件を`head`コマンドで表示するコマンドは以下．

`sort -k 3 -t '    ' -r -n  popular_names.txt | head -n 10`

各オプションの説明：
- `-k` : 何列目のフィールドでソートするか．今回は3．
- `-t` : フィールドごとの区切り文字．今回はタブ文字．
- `-r` : これにより降順になる．これがないと昇順になる．
- `-n` : 値を数字としてソートする．
